```python
import numpy as np
import cv2
import os
```

```python
# Define a function to convert images to matrices
def images_to_matrix(image_folder):
    i = 0 
    image_matrix = []
    for filename in os.listdir(image_folder):
        img = cv2.imread(os.path.join(image_folder, filename))
        if img is not None:
            # Resize and normalize the image
            print(i)
            img = img.astype('float32') / 255.0  # Normalize to [0, 1]
            image_matrix.append(img)
            i = i + 1
    return np.array(image_matrix)
```
